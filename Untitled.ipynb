```python
import yfinance as yf
from yahoo_fin import stock_info as si

def calculate_portfolio_stock_exposure(isin1, isin2, investment1, investment2):
    # Retrieve the holdings data for each ETF and calculate their total market values
    ticker1 = yf.Ticker(isin1)
    ticker2 = yf.Ticker(isin2)
    holdings_data1 = pd.DataFrame.from_records(ticker1.info["holdings"])
    holdings_data2 = pd.DataFrame.from_records(ticker2.info["holdings"])

    # Multiply the weight of each stock by the proportion of your investment in each ETF
    holdings_data1['exposure'] = holdings_data1['holdingPercent'] * investment1
    holdings_data2['exposure'] = holdings_data2['holdingPercent'] * investment2

    # Merge the exposures of each stock across both ETFs
    exposures = holdings_data1[['symbol', 'exposure']].merge(
        holdings_data2[['symbol', 'exposure']], on='symbol', how='outer'
    ).fillna(0)
    
    print(exposures)
    # Calculate the overall stock exposure of your portfolio
    total_exposure = exposures['exposure_x'] + exposures['exposure_y']
    portfolio_stock_exposure = ((total_exposure * 100) / (investment1 + investment2) ).to_dict()

    # Print the results
    print('Portfolio stock exposure:', portfolio_stock_exposure)

# Example usage
calculate_portfolio_stock_exposure('IE00B3XXRP09', 'IE00B3XXRP09', 5000, 5000)
```

```python
import yfinance as yf
import pandas as pd
from yahoo_fin import stock_info as si

def calculate_exposure(etfs):
    # create a dataframe to store the holdings and weights of all ETFs
    combined_df = pd.DataFrame()

    # iterate through each ETF and its investment
    for etf, investment in etfs.items():
        # get the holdings data for the current ETF
        ticker = yf.Ticker(etf)
        try:
            legalType = ticker.info['legalType']
            if legalType == "Exchange Traded Fund":
                print(f"INFO: {etf} is ISIN for ETF: {ticker.info['longName']}")
            else:
                print(f"INFO: {etf} is not an ETF, potentially corresponds to equity shares or stocks")
                pass
                
            holdings_data = pd.DataFrame.from_records(ticker.info["holdings"])
                
            # add the holdings and weights to the combined dataframe, using the investment amount as a weight
            holdings_data["investment_allocation"] = holdings_data["holdingPercent"] * investment
            combined_df = pd.concat([combined_df, holdings_data])
        except:
            print(f"WARNING: {etf} seems incorrect, please try again and reference Yahoo Finance for accurate ISIN")
        
    
    ticker_and_holding_name = combined_df[['symbol', 'holdingName']].drop_duplicates()
    stock_exposure = pd.DataFrame(combined_df.groupby("symbol").sum()["investment_allocation"])
    stock_exposure["Overall Exposure"] = (stock_exposure * 100 / sum(etfs.values()))
    final_output = stock_exposure.merge(
                                        ticker_and_holding_name, on='symbol', how='left'
                                    ).fillna(0)
    
    final_output = final_output.rename(columns={
                        'name': 'first_name', 
                        'investment_allocation': 'Amount Invested ($)',
                        'holdingName': 'Holding Name'
                        }
                  )

    
    return final_output.sort_values(by="Overall Exposure", ascending=False).reset_index(drop = True)
```

```python
etfs = {
    "IE00B3XXRP09": 5000,
    "IE00B6S2Z822": 7000,
    "IE00B53SZB19": 10000,
    "IE00BDZVHG35": 2000
}
```

```python
result = calculate_exposure(etfs)
```
